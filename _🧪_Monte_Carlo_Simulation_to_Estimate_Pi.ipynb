```python
import random
import matplotlib.pyplot as plt

# Monte Carlo simulation to estimate Pi
inside = 0
total = 1000
x_in, y_in, x_out, y_out = [], [], [], []

for _ in range(total):
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= 1:
        inside += 1
        x_in.append(x)
        y_in.append(y)
    else:
        x_out.append(x)
        y_out.append(y)

# Estimate of Pi
pi_estimate = 4 * inside / total
print("Estimated Pi:", pi_estimate)

# Plotting
plt.scatter(x_in, y_in, color='blue', label='Inside Circle', s=2)
plt.scatter(x_out, y_out, color='red', label='Outside Circle', s=2)
plt.title("Monte Carlo Pi Simulation")
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```
