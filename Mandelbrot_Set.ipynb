```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
width, height = 800, 800
x_min, x_max = -2.5, 1.5
y_min, y_max = -2, 2
```

```python
x = np.linspace(x_min, x_max, width)
y = np.linspace(y_min, y_max, height)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y
Z = np.zeros_like(C)
```

```python
max_iter = 100
diverge_time = np.zeros(C.shape, dtype=int)

for i in range(max_iter):
    Z = Z**2 + C
    mask = np.abs(Z) > 2
    diverge_time[mask & (diverge_time == 0)] = i
    Z[mask] = 2
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(diverge_time, cmap='magma', extent=(x_min, x_max, y_min, y_max))
plt.title("Mandelbrot Set")
plt.xlabel("Re")
plt.ylabel("Im")
plt.colorbar(label="Iterations to Diverge")
plt.show()
```
