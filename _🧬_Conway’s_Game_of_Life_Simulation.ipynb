```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Create random 10x10 grid of 0s and 1s
grid = np.random.choice([0, 1], size=(10, 10))

def count_neighbors(grid, row, col):
    total = 0
    for i in [-1, 0, 1]:
        for j in [-1, 0, 1]:
            if not (i == 0 and j == 0):
                total += grid[(row + i) % 10][(col + j) % 10]
    return total

def update(grid):
    new_grid = grid.copy()
    for i in range(10):
        for j in range(10):
            neighbors = count_neighbors(grid, i, j)
            if grid[i][j] == 1 and (neighbors < 2 or neighbors > 3):
                new_grid[i][j] = 0
            elif grid[i][j] == 0 and neighbors == 3:
                new_grid[i][j] = 1
    return new_grid

# Run the simulation for 10 generations
for generation in range(10):
    print(f"Generation {generation + 1}")
    plt.imshow(grid, cmap='binary')
    plt.title(f"Generation {generation + 1}")
    plt.show()
    grid = update(grid)
    time.sleep(0.5)
```
